```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data = pd.read_csv('train.csv')
```

```python
# Display basic information
print("Initial Data Preview:")
print(data[['GrLivArea', 'BedroomAbvGr', 'FullBath', 'SalePrice']].head())
```

```python
# Select relevant features and target
features = data[['GrLivArea', 'BedroomAbvGr', 'FullBath']]  # square footage, bedrooms, bathrooms
target = data['SalePrice']  # price
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

print("\nModel Performance:")
print(f"Mean Squared Error: {mse}")
print(f"R² Score: {r2}")
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=y_test, y=predictions, color="blue")
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual vs Predicted House Prices")
plt.grid(True)
plt.show()
```

```python
example_house = [[2000, 3, 2]]
example_house = pd.DataFrame({
    'GrLivArea': [2000],       # Square footage
    'BedroomAbvGr': [3],       # Number of bedrooms
    'FullBath': [2]            # Number of bathrooms
})

predicted_price = model.predict(example_house)
print(f"Predicted House Price: ${predicted_price[0]:,.2f}")
```
